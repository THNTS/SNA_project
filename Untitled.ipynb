Degree distribution

```python


import collections

#draw degree distribution
components = nx.connected_components(G)
largest_component = max(components, key=len)
G_large = G.subgraph(largest_component)
def plot_degree_distribution(G_large):
    degree_sequence = sorted([d for n, d in G_large.degree()], reverse=True)  # degree sequence
    degreeCount = collections.Counter(degree_sequence)
    deg, cnt = zip(*degreeCount.items())
    fig, ax = plt.subplots()
    plt.title("Degree Distribution")
    plt.bar(deg, cnt)
  
plot_degree_distribution(G_large)

```

```python
import os
import glob
import pandas as pd
# os.chdir("/directory")

file_extension = '.csv'
all_filenames = [i for i in glob.glob(f"/Users/daniilbabin/DS/JobDemand/Data/*{file_extension}")]
combined_csv_data = pd.concat([pd.read_csv(f) for f in all_filenames])
```

```python
test = combined_csv_data.sample(n=3000, random_state=1)
```

```python
texts = test['Job Description'].tolist()
```

```python
import nltk

sentences = []

for text in texts:
    new = []
    paragraphs = [p for p in text.split('\n') if p]
    for paragraph in paragraphs:
        new.extend(nltk.sent_tokenize(paragraph))
    sentences.append(new)
```

```python
from nltk.tokenize import word_tokenize
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('wordnet')

def clear_sentence(sentence):
    table = str.maketrans(string.punctuation, ' '*len(string.punctuation))
    stripped = sentence.translate(table)
    tokens = word_tokenize(stripped)
    tokens = [w.lower() for w in tokens]
    words = [word for word in tokens if word.isalpha()]
    stop_words = set(stopwords.words('english'))
    porter = PorterStemmer()
    wnl = WordNetLemmatizer()
    lemmatized = [wnl.lemmatize(word) for word in words]
    stemmed = [porter.stem(word) for word in words]
    return lemmatized
```

```python
cleared = []
for text in sentences:
    new = []
    for sentence in text:
        clear = clear_sentence(sentence)
        new.extend(clear)
    cleared.append(new)
```

```python
cleared_new = [set(ele) for ele in cleared]
```

```python
from collections import Counter
count_reqs = [list(a) for a in reqs]
c = Counter(c for clist in count_reqs for c in clist)
```

```python
import itertools

df1 = pd.DataFrame({"subject": list(list_of_req)})
df2 = pd.DataFrame({"subject_list": count_reqs})

# Create a new dataframe with one column that has the possible two by two combinations from `df1`
df_combinations = pd.DataFrame({'combination': list(itertools.combinations(df1.subject, 2))})
```

```python
for index, row in df_combinations.iterrows():

    df_combinations.at[index, "number of occurrences"] = df2["subject_list"].apply(lambda x: all(i in x for i in row['combination'])).sum()
```
